a. From the given information, we can see that $$xn+1 = xn − xnΔt + √2Δt ξn $$where $$ξn ∼ N(0, 1)$$ for n = 0, 1, . . .. This means that xn+1 is normally distributed with mean xn − xnΔt and variance 2Δt. Therefore, x1 is normally distributed with mean x0 − x0Δt and variance 2Δt. Similarly, x2 is normally distributed with mean x1 − x1Δt and variance 2Δt.

b. This code uses the numpy and matplotlib libraries to generate random numbers and plot the results. You can run this code to generate histograms of the empirical distributions at time t = 0, t = 0.5, t = 1 and t = 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
dt = 0.01
x0 = 2
T = 10
N = int(T/dt)
n_simulations = 10000

# Initialize array to store results
X = np.zeros((n_simulations, N+1))
X[:, 0] = x0

# Run simulations
for i in range(n_simulations):
    for n in range(N):
        xi = np.random.normal(0, 1)
        X[i, n+1] = X[i, n] - X[i, n]*dt + np.sqrt(2*dt)*xi

# Plot histograms
times = [0, 0.5, 1, 10]
for t in times:
    plt.figure()
    plt.hist(X[:, int(t/dt)], bins=50)
    plt.title(f't = {t}')
    plt.show()
```
